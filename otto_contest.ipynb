## Отнормируем данные
и посмотрим на распределение значений признаков

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
sct = StandardScaler()
#некоторые колонки нормировать абсолютно бессымысленно
columns = [u'id', u'target']
columns_for_scaling = [s for s in train_df.columns.tolist()\
                       if s not in columns]

scaled_train_df = train_df.copy()
scaled_train_df[columns_for_scaling] = sct.fit_transform(train_df[columns_for_scaling])
```

```python
#посмотрим распределение признаков
xs = [scaled_train_df[feat].value_counts().sort_index().index.tolist()\
      for feat in columns_for_scaling]

ys = [scaled_train_df[feat].value_counts().sort_index().values\
      for feat in columns_for_scaling]
```

Признаков много, и трудно что-то сказать, глядя на графики.

## Разобъем данные и построим модели

Вспомним, что количество элементов в классах сильно различается.

```python
classes_count = scaled_train_df.groupby('target')['id'].count()
classes_count
```

```python
#посчитаем, во сколько раз каждого класса меньше, чем максимального
max_num = max(classes_count)
min_num = min(classes_count)
k_for_classes = [max_num / i for i in classes_count]
m_for_classes = [1.*min_num / i for i in classes_count]
```

Порядок данных в таблице нам не важен, поэтому можем смело мешать данные и делать кросс-валидацию.
Есть несколько способов борьбы с дисбалансом классов.

Способ 1. Разбивать выборки более-менее сбалансированно.
Возможно, что данных будет мало, но зато меньше вероятность переобучиться.

Способ 2. Раздуть выборки, продублировав несколько раз строки из маленьких классов.
Данных станет больше, но есть шанс переобучится. Можно сначала разбить выборку на части, а затем "раздувать" ее, чтобы избежать попадания одних и тех же строк в train и test.

Способ 3. Возможно, следует сохранить дисбаланс, если в данных для проверки так же имеется дисбаланс.

В качестве метрики качества будем использовать то, что предлагают организаторы соревнования - log loss.

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold, KFold, train_test_split
from sklearn.learning_curve import validation_curve
from sklearn.linear_model import LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
import xgboost
```

Отберем признаки и метки классов из датафреймов

```python
#3 способ
#для нормированных данных (для моделей, зависящих от масштаба данных)
scaled_train_df.index = scaled_train_df['id']
X_sc = scaled_train_df[columns_for_scaling]

#target теперь номер класса от 0 до 8, а не Class_k
Y_sc = scaled_train_df['target'].apply(lambda x: int(x[-1])-1)

#для ненормированных данных
train_df.index = train_df['id']
X = train_df[columns_for_scaling]
Y = train_df['target'].apply(lambda x: int(x[-1])-1)
```

"Раздуем" исходный датафрейм и сделаем для него то же самое

```python
#2 способ
class_frames = [train_df[train_df['target'] == 'Class_'+str(i)] for i in range(1,10)]
frames_to_concat = [cf for cf,k in zip(class_frames,k_for_classes) for i in range(k)]
b_train_df = pd.concat(frames_to_concat)
```

```python
b_train_df.groupby('target')['id'].count()
```

```python
X_b = b_train_df[columns_for_scaling]
Y_b = b_train_df['target'].apply(lambda x: int(x[-1])-1)

sct = StandardScaler()
b_scaled_train_df = b_train_df.copy()
b_scaled_train_df[columns_for_scaling] = sct.fit_transform(b_train_df[columns_for_scaling])
X_b_sc = b_scaled_train_df[columns_for_scaling]
Y_b_sc = Y_b 
```

Отбросим часть данных, для того, чтобы датафрейм был сбалансированным

```python
#1 способ
frames_to_concat = [train_test_split(cf,test_size = m, random_state = 42)[1]\
                    if m < 1 else cf for cf,m in zip(class_frames,m_for_classes)]
s_train_df = pd.concat(frames_to_concat)

X_s = s_train_df[columns_for_scaling]
Y_s = s_train_df['target'].apply(lambda x: int(x[-1])-1)

sct = StandardScaler()
s_scaled_train_df = s_train_df.copy()
s_scaled_train_df[columns_for_scaling] = sct.fit_transform(s_train_df[columns_for_scaling])
X_s_sc = s_scaled_train_df[columns_for_scaling]
Y_s_sc = Y_s 
```

```python
#освободим память
del(train_df,b_train_df,scaled_train_df,class_frames,frames_to_concat, \
    b_scaled_train_df, s_scaled_train_df,s_train_df)
```

```python
#разобъем данные 
#c сохранением пропорций классов:
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
#и без
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

Обучим несколько моделей на двух выборках и сравним результаты

## Logistic Regression

С исходным дисбалансом классов

```python
LR = LogisticRegressionCV(penalty='l2', n_jobs=-1)
cv = skf.split(X,Y)
LR_scores = cross_val_score(LR, X, Y, scoring='neg_log_loss', cv=cv)
print('Logistic Regression log loss: {0:.3f} ± {1:.3f}'.format(LR_scores.mean(), LR_scores.std()) )
```

На "раздутых" данных

```python
LR = LogisticRegressionCV(penalty='l2', n_jobs = -1)
cv = kf.split(X_b,Y_b)
LR_scores_b = cross_val_score(LR, X_b, Y_b, scoring='neg_log_loss', cv=cv)
print('Logistic Regression log loss: {0:.3f} ± {1:.3f}'.format(LR_scores_b.mean(), LR_scores_b.std()) )
```

На части данных со сбалансированными классами

```python
LR = LogisticRegressionCV(penalty='l2', n_jobs = -1)
cv = kf.split(X_s,Y_s)
LR_scores_s = cross_val_score(LR, X_s, Y_s, scoring='neg_log_loss', cv=cv)
print('Logistic Regression log loss: {0:.3f} ± {1:.3f}'.format(LR_scores_s.mean(), LR_scores_s.std()) )
```

## Random Forest

С исходным дисбалансом классов

```python
rf = RandomForestClassifier(n_jobs=-1)

cv = skf.split(X, Y)
prange = np.arange(20,270,50,dtype=int)
rf_train_scores, rf_cv_scores = validation_curve(rf, X, Y, param_name = 'n_estimators',\
                                           param_range=prange, cv = cv, scoring='neg_log_loss')
plt.title('max_depth = None')
plt.xlabel('n_estimators')
plt.ylabel('score')
plt.plot(prange, np.mean(rf_cv_scores, axis=1))
plt.show()
```

```python
cv = skf.split(X, Y)
rf = RandomForestClassifier(n_jobs=-1, n_estimators=100)
rf_scores = cross_val_score(rf, X, Y, scoring='neg_log_loss', cv=cv)
print('Random Forest log loss: {0:.3f} ± {1:.3f}'.format(rf_scores.mean(), rf_scores.std()) )
```

На "раздутых" данных

```python
rf = RandomForestClassifier(n_jobs=-1)

cv = kf.split(X_b, Y_b)
prange = np.arange(20,270,50,dtype=int)
rf_train_scores, rf_cv_scores = validation_curve(rf, X_b, Y_b, param_name = 'n_estimators',\
                                           param_range=prange, cv = cv, scoring='neg_log_loss')
plt.title('max_depth = None')
plt.xlabel('n_estimators')
plt.ylabel('score')
plt.plot(prange, np.mean(rf_cv_scores, axis=1))
plt.show()
```

```python
cv = kf.split(X_b, Y_b)
rf = RandomForestClassifier(n_jobs=-1, n_estimators=100)
rf_scores = cross_val_score(rf, X_b, Y_b, scoring='neg_log_loss', cv=cv)
print('Random Forest log loss: {0:.3f} ± {1:.3f}'.format(rf_scores.mean(), rf_scores.std()) )
```

На части данных со сбалансированными классами

```python
rf = RandomForestClassifier(n_jobs=-1)

cv = kf.split(X_s, Y_s)
prange = np.arange(20,270,50,dtype=int)
rf_train_scores_s, rf_cv_scores_s = validation_curve(rf, X_s, Y_s, param_name = 'n_estimators',\
                                           param_range=prange, cv = cv, scoring='neg_log_loss')
plt.title('max_depth = None')
plt.xlabel('n_estimators')
plt.ylabel('score')
plt.plot(prange, np.mean(rf_cv_scores_s, axis=1))
plt.show()
```

```python
cv = kf.split(X_s, Y_s)
rf = RandomForestClassifier(n_jobs=-1, n_estimators=70)
rf_scores_s = cross_val_score(rf, X_s, Y_s, scoring='neg_log_loss', cv=cv)
print('Random Forest log loss: {0:.3f} ± {1:.3f}'.format(rf_scores_s.mean(), rf_scores_s.std()) )
```

## XGBoost

```python
#xgboost
param = {}
#param['updater'] = 'grow_gpu'
param['booster'] = 'gbtree'
param['nthread'] = 16

param['objective'] = 'multi:softprob'
param['eval_metric'] = 'mlogloss'
param['num_class'] = 9

param['eta'] = 0.05
param['max_depth'] = 5
param['subsample'] = 0.5
param['rate_drop'] = 0.5

numround = 300
```

С исходным дисбалансом классов

```python
Xdata_cv = xgboost.DMatrix(data = X, label = Y)

plst = list(param.items())
bst_cv_d = xgboost.cv(plst, Xdata_cv, numround, nfold = 5, stratified=True,\
                      verbose_eval=50, seed=42, show_stdv=True)
```

## Результаты

Признаков много, нет их описания, поэтому как то интерпретировать их и сделать отбор признаков трудно.
Распределения признаков похожи, очень много нулей. Гипотеза о земене NaN нулями не подтвердилась. Все значения целые, а средние значения нет, поэтому заменить NaN средними так же не могли.

Как и предполагалось, если сбалансировать классы "раздувая" датасет, то модели переобучаются.
RF показал ~0.21 log loss на кроссвалидации предложенным методом (топ1 на лидерборде ~0.3). Исполбзовать kfold с раздутой выборкой не совсем корректно, так как в отложенном фолде встречаются элементы из тренировочных фолдов, поэтому cross_val_score получается сильно завышен.

Если же сбалансировать классы, выкидывая часть данных, то данных становится слишком мало, и обобщающая способность модели оставляет желать лучшего. Это видно по оценкам LR, RF и разрыву между test-mlogloss и train-mlogloss для xgboost.

Оптимальным решением оказалось оставить данные как есть, и обучаться на них. Возможно, можно улучшить результат, убирая меньше данных (например, чтобы отношения размера самого большого класса к самому маленькому было не больше 2.5-3), или немного "раздувая" датасет.

LR очень медленно обучается и показывает худшие результаты. Лучший результат на предложененой кросс-валидации и метрике качества показывает xgboost.
